```python
#X = np.array([[1, 2], [1, 4], [1, 0],[4, 2], [4, 0], [4, 4],[4, 5], [0, 1], [2, 2],[3, 2], [5, 5], [1, -1]])
n=len(X)
kmeans = MiniBatchKMeans(n_clusters=int(n/div),
        random_state=0,
        batch_size=int(n/div))
for batch in range(div):
    print(batch)
    print()
    kmeans = kmeans.partial_fit(X[batch*int(n/div):(batch+1)*int(n/div)])
print(kmeans.cluster_centers_)
```

```python
closest, _ = pairwise_distances_argmin_min(kmeans.cluster_centers_, X)
```

```python
from shutil import copyfile

seen = set()
count = 0
for i, el in enumerate(closest):
    if el not in seen:
        print(el)
        try:
            copyfile('full_dataset/datasetlvl'+str(lvl)+'/dataset/'+str(el),'dataset'+str(div)+'/datasetlvl'+str(lvl)+'/dataset/'+str(count))
            copyfile('full_dataset/datasetlvl'+str(lvl)+'/features/'+str(el),'dataset'+str(div)+'/datasetlvl'+str(lvl)+'/features/'+str(count))
            count +=1
            seen.add(el)
        except:
            print(el, " Not founded")
```
